# 17. Какой актер принес меньше всего прибыли в 2012 году?
Варианты ответов:
1. Nicolas Cage
2. Danny Huston
3. Kirsten Dunst
4. Jim Sturgess
5. Sami Gayle

```python
dg = data[ data['release_year'] == 2012 ]

c = Counter()

for index, row in dg.iterrows():
    names = row['cast'].split('|')
    for name in names:
        c[name] += row['прибыль'] / len(names)
        
        
display(c.most_common()[:-2:-1])

answer_ls.append(3)
```

# 31. Названия фильмов какой студии в среднем самые длинные по количеству символов?
Варианты ответа:
1. Universal Pictures (Universal)
2. Warner Bros
3. Jim Henson Company, The
4. Paramount Pictures
5. Four By Two Productions

```python
data['длина_названия'] = data['original_title'].map(lambda x: len(x))

companies = data['production_companies'].str.cat(sep='|')
companies = pd.Series(companies.split('|')).value_counts()

for i in companies.index:
    companies[i] = data['длина_названия'][data['production_companies'].map(lambda x: True if i in x else False)].mean()
    
display(pd.DataFrame(companies).sort_values(0, ascending=False))

answer_ls.append(5)
```

# 36. У какого из режиссеров выше вероятность выпустить фильм в прибыли? (5 баллов)101
Варианты ответа:
1. Quentin Tarantino
2. Steven Soderbergh
3. Robert Rodriguez
4. Christopher Nolan
5. Clint Eastwood

```python
ds = pd.Series(data['director'].str.cat(sep='|').split('|')).unique()

ds = pd.DataFrame(columns=['director'], data=ds)
ds['процент'] = None

for i in ds.index:
    dm = data[ data['director'].str.contains(ds.iloc[i]['director']) ]
    ds.loc[i,'процент'] = len(dm[ dm['прибыль'] > 0 ]) * 100 / len(dm)
    

display(ds[ ds['director'].isin(['Quentin Tarantino', 'Steven Soderbergh', 'Robert Rodriguez', 'Christopher Nolan', 'Clint Eastwood'])].sort_values('процент'))


answer_ls.append(4)
```
